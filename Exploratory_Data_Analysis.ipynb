```python
house_prices['price_per_sqft'] = house_prices['price'] / house_prices['room_num']
kc_data['price_per_sqft'] = kc_data['price'] / kc_data['sqft_living']
```

```python
 plt.figure(figsize=(10, 6))
# Check if 'long' exists, if not try 'longitude', otherwise use a different column
x_col = "long" if "long" in kc_data.columns else ("longitude" if "longitude" in kc_data.columns else "sqft_living")
# Check if 'lat' exists, if not try 'latitude', otherwise use a different column
y_col = "lat" if "lat" in kc_data.columns else ("latitude" if "latitude" in kc_data.columns else "price")
sns.scatterplot(x=kc_data[x_col], y=kc_data[y_col], hue=kc_data["price"], palette="coolwarm", alpha=0.6)
plt.title("House Prices by Location")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Simple Regression on house_prices
features_hp = ['room_num', 'crime_rate', 'n_hot_rooms']
X_hp = house_prices[features_hp]
y_hp = house_prices['price']

model_hp = LinearRegression()
model_hp.fit(X_hp, y_hp)

print("House Price Model Coefficients:", model_hp.coef_)
print("House Price Intercept:", model_hp.intercept_)

# For KC dataset
features_kc = ['sqft_living', 'bedrooms', 'bathrooms']
X_kc = kc_data[features_kc]
y_kc = kc_data['price']

model_kc = LinearRegression()
model_kc.fit(X_kc, y_kc)

print("KC Dataset Model Coefficients:", model_kc.coef_)
print("KC Dataset Intercept:", model_kc.intercept_)
```
